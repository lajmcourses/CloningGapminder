## 3. Pirminių duomenų šaltinių analizė


### 3.1 Pasaulio šalys

```python
# Data reading from csv

df_countries = pd.read_csv("data/countries.csv") 

df_countries.head()
```

```python
# Stulpelių pervadinimas: 'ISO3':'Country Code', 'Country':'CountryName'
df_countries.rename({
    "ISO3": "Country Code",
    "Country": "Country Name"
}, axis="columns", inplace=True)

# Stulpelių pašalinimas
df_countries.drop(columns=["ISO Codes", "ISO2"], inplace=True)
```

```python
df_countries.head()
```

```python
# Duomenų kiekis

print("Number of rows:", df_countries.shape[0])
print("Number of columns:", df_countries.shape[1])
```

```python
# Bazinė duomenų statistika

df_countries.describe()
```

### 3.2 Gimstamumas

```python
# Data reading from csv

df_fertility = pd.read_csv("data/wb_fertility_rate.csv") 
df_fertility.head()
```

## Duomenų redagavimas: šalių duomenys, kurių nėra df_contries lentelėje, pašalinimas

```python
# Stulpelių pašalininas: Indicator Name, Indicator Code

df_fertility.drop(columns=["Indicator Name", "Indicator Code"], inplace=True)
df_fertility.head()
```

```python
df_fertility = pd.merge(df_countries[["Country Code"]], df_fertility, on="Country Code", how="left")
```

```python
# Duomenų kiekis

print("Number of rows:", df_fertility.shape[0])
print("Number of columns:", df_fertility.shape[1])
```

```python
df_fertility.head()
```

```python
df_fertility.describe()
```

```python
df_fertility.describe(include=object)
```

### 3.3 Gyvenimo trukmė

```python
# Data reading from csv

df_life_expectancy = pd.read_csv("data/wb_life-expectancy.csv") 

df_life_expectancy.head()
```

## Duomenų redagavimas: šalių duomenys, kurių nėra df_contries lentelėje, pašalinimas

```python
df_life_expectancy = pd.merge(df_countries[["Country Code"]], df_life_expectancy, on="Country Code", how="left")
```

```python
# Duomenų kiekis

print("Number of rows:", df_life_expectancy.shape[0])
print("Number of columns:", df_life_expectancy.shape[1])
```

```python
# Skaitmeninių duomenų statistika
df_life_expectancy.describe()
```

```python
# Kategorinių duomenų statistika
df_life_expectancy.describe(include=object)
```
